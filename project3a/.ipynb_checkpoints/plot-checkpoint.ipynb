```python
import re
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ns = [1, 2, 4, 8, 14]
locks = ['TAS', 'TTAS', 'Mutex', 'ALock', 'CLH']
cols = ['Trial #', 'n', 'Serial'] + locks
```

```python
def strip_floats(line):
  return [float(elm) for elm in re.findall(r'[0-9]+\.[0-9]+', line)]
```

```python
def to_df(fname):
  with open(fname, 'r') as f:
    raw = f.read()
  trials = raw.split('#####')[1:]
  
  clean = []
  for t, trial in enumerate(trials):
    clean.append(
      [t + 1, 1] + strip_floats(''.join(trial.split('###')[0:2])))
    for i, n in enumerate(ns[1:]):
      clean.append(
        [t + 1, n, 0] + strip_floats(trial.split('###')[i + 2]))
      
  df = pd.DataFrame(columns=cols)
  for line in clean:
    mapped = {key: val for key, val in zip(cols, line)}
    df = df.append(mapped, ignore_index=True)
  df = df.astype({'Trial #': int, 'n': int})
  return df
```

```python
def clean_df(df):
  df2 = df.groupby(['n']).median().drop('Trial #', axis=1)
  df2 = big / df2
  df2.insert(0, 'n', ns)
  return df2
```

```python
def to_overhead_df(df2):
  overhead_df = df2.loc[df2['n'] == 1] / serial_throughput
  overhead_df = overhead_df.drop(['n', 'Serial'], axis=1)
  overhead_df = overhead_df.transpose()
  return overhead_df
```

```python
def to_scale_df(df2):
  scale_df = df2 / serial_throughput
  scale_df = scale_df.drop(['n', 'Serial'], axis=1)
  return scale_df
```

```python
big = 10000000
df = to_df('result/BIG_10M.txt')
df2 = clean_df(df)
serial_throughput = df2.loc[df2['n'] == 1, 'Serial'].values[0]
print(round(serial_throughput))
overhead_df = to_overhead_df(df2)
scale_df = to_scale_df(df2)
```

```python
plt.figure()
scale_df.plot(title='Lock Scaling with BIG={}'.format(big))
plt.xticks(ns)
plt.xlabel('n')
plt.ylabel('Increments per Second')
plt.show()
```

```python
zoomed = scale_df[1:]
plt.figure()
zoomed.plot(title='Zoomed: Lock Scaling with BIG={}'.format(big))
plt.xticks(ns[1:])
plt.xlabel('n')
plt.ylabel('Increments per Second')
plt.show()
```
