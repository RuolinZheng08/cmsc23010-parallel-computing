```python
import re
import csv
import itertools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def strip_nums(line):
  return re.findall(r'-?[0-9]*\.?[0-9]+', line)
```

```python
# Trial format for Dispatcher Rate
# # Trial 1

# * parallel no load
# Parameters: -m 2000, -a 0.090000, -r 0.010000, -h 0.900000, -w 1, -s 0; -n 13, -t (null)
# Total Time Handling Packets: 2000.922974
# Total Packets: 2650896
```

```python
# Trial format for Parallel Overhead
# # Trial 1

# ** Mostly Reads

# * serial
# Parameters: -m 2000, -a 0.090000, -r 0.010000, -h 0.900000, -w 4000, -s 0; -n -1, -t (null)
# Total Time Handling Packets: 2000.098999
# Total Packets: 406367

# Trial format for Speedup
# # Trial 1 with n=1 and h=0.5

# ** Mostly Reads

# * serial
# Parameters: -m 2000, -a 0.090000, -r 0.010000, -h 0.500000, -w 4000, -s 0; -n -1, -t (null)
# Total Time Handling Packets: 2000.118042
# Total Packets: 407111
```

```python
with open('result/master_max_bkt_size_4.txt', 'r') as f:
  raw = f.read()
tests = raw.split('###')[1:]
dispatcher, overhead, speedup = tests
```

```python
def parse_dispatcher_rate(dispatcher):
  res = []
  trials = dispatcher.split('#')[1:]
  for t in trials:
    num = int(re.findall(r'Total Packets: ([0-9]+)', t)[0])
    res.append(num)
  return np.array(res) / 2000
```

```python
np.median(parse_dispatcher_rate(dispatcher))
```

```python
cols = ['T', 'serial reads', 'serial writes', 
        'finelock reads', 'finelock writes', 
       'optimistic reads', 'optimistic writes']
def parse_overhead(overhead):
  df = pd.DataFrame(columns=cols)
  trials = overhead.split('#')[1:]
  for i, t in enumerate(trials):
    t_dict = {'T': i + 1}
    chunks = t.split('**')[1:]
    read, write = chunks
    rlines, wlines = read.split('*')[1:], write.split('*')[1:]
    for rl, wl in zip(rlines, wlines):
      rdata = re.findall(' (.*)\n', rl)[0], int(strip_nums(rl)[-1])
      wdata = re.findall(' (.*)\n', wl)[0], int(strip_nums(wl)[-1])
      t_dict[rdata[0] + ' reads'] = rdata[1]
      t_dict[wdata[0] + ' writes'] = wdata[1]
    df = df.append(t_dict, ignore_index=True)
  return df
```

```python
parse_overhead(overhead)
```

```python
def strip_stats(line):
  prog = re.findall(' (.*)\n', line)[0]
  num_workers = int(re.findall('-n (-?[0-9]+)', line)[0])
  hit_rate = float(re.findall('-h ([0-9].?[0-9]*)', line)[0])
  pkts = int(strip_nums(line)[-1])
  return prog, num_workers, hit_rate, pkts
```

```python
strip_stats(line)
```

```python
cols = ['T', 'n', 'hit rate', 'serial reads', 'serial writes', 
        'finelock reads', 'finelock writes', 
       'optimistic reads', 'optimistic writes']
def parse_speedup(speedup):
  df = pd.DataFrame(columns=cols)
  trials = speedup.split('#')[1:]
  for t in trials:
    t_no = re.findall(r'Trial ([0-9])', t)[0]
    t_dict = {'T': t_no}
    chunks = t.split('**')[1:]
    read, write = chunks
    rlines, wlines = read.split('*')[1:], write.split('*')[1:]
    for rl, wl in zip(rlines, wlines):
      rstats = strip_stats(rl)
      wstats = strip_stats(wl)
      t_dict[rstats[0] + ' reads'] = rstats[3]
      t_dict[wstats[0] + ' writes'] = wstats[3]
      t_dict['n'] = rstats[1]
      t_dict['hit rate'] = rstats[2]
    df = df.append(t_dict, ignore_index=True)
  return df
```

```python
parse_speedup(speedup)
```

Clean Data

```python
overhead_df = pd.read_csv('result/overhead.csv')
speedup_df = pd.read_csv('result/speedup.csv')
```

```python
temp = overhead_df.median()
flr = temp['finelock reads'] / temp['serial reads']
fwr = temp['finelock writes'] / temp['serial writes']
olr = temp['optimistic reads'] / temp['serial reads']
owr = temp['optimistic writes'] / temp['serial writes']
overhead_clean = pd.DataFrame.from_dict(
  {'finelock reads': flr, 'finelock writes': fwr, 
   'optimistic reads': olr, 'optimistic writes': owr}, 
  orient='index').transpose()
```

```python
print(overhead_clean.to_latex(index=False))
```

```python
temp = speedup_df.groupby(['n', 'hit rate']).median()
```

```python
htlist = ['finelock', 'optimistic']
metlist = ['reads', 'writes']
for ht, met in itertools.product(htlist, metlist):
  temp['{}-{}'.format(ht, met)] = \
  temp[ht + ' ' + met] / temp['serial ' + met]
```

```python
result = temp.drop(
  columns=['T', 'serial reads', 'serial writes', 'finelock reads',
          'finelock writes', 'optimistic reads', 'optimistic writes'])
result = result.reset_index()
```

```python
for hr in [0.5, 0.75, 0.9, 0.99]:
  subset = result.loc[result['hit rate'] == hr].drop(columns='hit rate')
  subset.plot(x='n', title='Hit Rate = {}'.format(hr))
  print(subset.to_latex(index=False))
```
